```python
import pandas as pd
```

```python
#1. Імпортуй датасет у Pandas
# шлях до датасету на гугл диску
file_id = "12pYkauOit0YEfNocUfc4DgRGcyuiIXH6"
url = f"https://drive.google.com/uc?id={file_id}"

df = pd.read_csv(url)

#https://drive.google.com/file/d/12pYkauOit0YEfNocUfc4DgRGcyuiIXH6/view?usp=sharing
```

```python
#2. Приведи Date до типу datetime
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
#перевірка рядків, які не конвертувалися
print("Дати, які не конвертувалися:")
print(df[df['Date'].isna()])

#імпортуємо оригінальний датасет
df_in = pd.read_csv(url)
#окремо обробляємо дату з відмінним форматом
df.loc[26, 'Date'] = pd.to_datetime(df_in.loc[26, 'Date'], format='%Y%m%d', errors='coerce')
print("\nЗначення в рядку 26 після виправлення:")
print(df.loc[26, 'Date'])
print("\nТип даних стовпця Date:")
print(df['Date'].dtype)
```

```python
#3. Перевір структуру таблиці: shape, columns, dtypes, info(), head()
print (f"Розмір таблиці: {df.shape}")
print(f"\nМістить колонки: {df.columns}")
print(f"\nЗ наступними типами:\n{df.dtypes}")
print("\nОгляд структури:")
df.info()
print("\nПерші 5 рядків:")
df.head()
```

```python
#4. Перевір наявність пропусків, виведи частку
missing_info = pd.DataFrame({
    'Кількість пропусків': df.isna().sum(),
    'Частка пропусків (%)': df.isna().mean()*100
})

print(missing_info)
```

```python
#5. Розрахуй статистики для Calories: mean, median, min, max, std, var
print(f"Середнє значення для Calories: {df['Calories'].mean()}")
print(f"Медіана для Calories: {df['Calories'].median()}")
print(f"Мінімальне значення для Calories: {df['Calories'].min()}")
print(f"Максимальне значення для Calories: {df['Calories'].max()}")
print(f"Стандартне відхилення для Calories: {df['Calories'].std()}")
print(f"Дисперсія для Calories: {df['Calories'].var()}")
```

Середнє значення 304.68 трохи вище за медіану 291.2
Це вказує на те, що розподіл спалених калорій є злегка скошеним вправо (тобто є кілька тренувань із дуже високим показником калорій)

Діапазон значень досить широкий (від 195.1 до 479.0), а стандартне відхилення (66.00) показує помірну варіативність: більшість тренувань мають показники, що відхиляються від середнього приблизно на +/- 66 калорій.

```python
#6. Розрахуй Q1, Q3, IQR і знайди викиди
# Розраховуємо квартилі
q1 = df['Calories'].quantile(0.25)
q3 = df['Calories'].quantile(0.75)
print(f"Q1 = {q1}")
print(f"Q3 = {q3}")
# Міжквартильний розмах
iqr = q3 - q1
print(f"IQR = {iqr}")
# Знаходимо межі викидів
low = q1 - 1.5*iqr
high = q3 + 1.5*iqr
print(f"Нижня межа: {low}")
print(f"Верхня межа: {high}")
print(f"Нормальний діапазон: {low:.0f} - {high:.0f} калорій")

# Знаходимо всі викиди
outliers = df[(df["Calories"] < low) | (df["Calories"] > high)]

print(f"Кількість викидів: {outliers.shape[0]}")
print(f"Відсоток викидів: {(outliers.shape[0] / df.shape[0]) * 100:.1f}%")
```

Перевірка даних показала, що у стовпці Calories немає жодного викиду, оскільки жодне значення не є меншим за 111 і не є більшим за 484.

```python
#7. Створи колонку is_calories_missing, заповни пропуски середнім значенням
#створюємо нову колонку із булевим значенням True/False в залежності від наявності пропуску
df['is_calories_missing'] = df['Calories'].isna()
#заповнюємо пропуски у Calories середнім значенням
df['Calories'].fillna(df['Calories'].mean(), inplace = True)
#виводимо результат
print(df[['Calories', 'is_calories_missing']])
```

Після заповнення пропусків у Calories середнім значенням, попередній код покаже, що у нас з'явився 1 викид

```python
#8. Створи колонку pulse_range = Maxpulse - Pulse
df['pulse_range'] = df['Maxpulse'] - df['Pulse']
#виводимо результат
print(df[['Maxpulse', 'Pulse', 'pulse_range']])
```

```python
#9. Створи категорії за Duration: <30 30–60 >60
for i in range(0,31,1):
  if df.loc[i, 'Duration'] < 30:
    df.loc[i, 'Duration_Category'] = '<30'
  elif df.loc[i, 'Duration'] <= 60:
    df.loc[i, 'Duration_Category'] = '30-60'
  else:
    df.loc[i, 'Duration_Category'] = '>60'
#виводимо результат
print(df[['Duration', 'Duration_Category']])
```

```python
#10. Побудуй агреговану таблицю з розрахунком середнього Pulse і Calories для кожної групи
duration_info = df.groupby('Duration_Category').agg({
    'Pulse' : 'mean',
    'Calories' : 'mean'
}).reset_index()

duration_info.columns = ['duration category', 'avg_pulse', 'avg_calories']
print(duration_info)
```
