Завдання 2. Аналіз повноти відповідей респондентів

```python
#створюємо множину із унікальних запитань зі схеми
schema_questions = set(schema_copy['qname'].unique())
#print(schema_questions)

#створюємо множину із назвами колонок основних даних
public_cols = set(public_copy.columns)
#print(public_cols)

#знаходимо інтерсекцію (перетин) двох множин, це покаже, які запитання зі схеми дійсно присутні у наших основних даних
survey_questions = schema_questions & public_cols
print(f"Основні дані містять {len(survey_questions)} із {len(schema_questions)} запитань зі схеми!")

missing_questions = schema_questions - public_cols

print(f"Відсутні запитання ({len(missing_questions)}):")
print(missing_questions)
print("="*50)

full_response_df = public_copy.dropna(subset=survey_questions)
print(f"Кількість респондентів, які відповіли на усі питання: {full_response_df.shape[0]}")
```

```python
#рахуємо, на скільки питань (із survey_questions) відповів кожен респондент
answered_count = public_copy[list(survey_questions)].count(axis=1)

answers_mean = answered_count.mean()
print(f"Середня кількість відповідей: {answers_mean:.2f}")
answers_median = answered_count.median()
print(f"Медіанна кількість відповідей: {answers_median:.2f}")

#графік
plt.figure(figsize=(10, 6))
plt.hist(answered_count, bins=50, color='skyblue', edgecolor='black')
plt.title('Скільки питань заповнюють люди?', fontweight = 'bold', fontsize = 14)
plt.xlabel('Кількість відповідей', fontweight = 'bold')
plt.ylabel('Кількість людей', fontweight = 'bold')
plt.axvline(answers_mean, color='green', linestyle='dashed', linewidth=2, label=f'Середнє ({int(answers_mean)} питань)')
plt.axvline(answers_median, color='red', linestyle='dashed', linewidth=2, label=f'Медіана ({int(answers_median)} питань)')
plt.legend()
plt.show()
```

Висновок: Розрахунки підтверджені візуалізацією - жоден респондент не відповів на усі питання анкети, але більшість респондентів відповіли більше ніж на половину питань анкети

Завдання 4. Аналіз віддаленої роботи


```python
#функція для пошуку назви колонки за значенням
def find_columns_by_value(df, search_value, case_sensitive = False):
  found_cols = []
  #шукаємо у колонках із текстом
  text_cols = df.select_dtypes(include = 'object').columns

  for col in text_cols:
    if df[col].str.contains(search_value, case = case_sensitive, na = False).any():
      found_cols.append(col)

  return found_cols
```

```python
keyword = "remote"
columns_with_remote = find_columns_by_value(public_copy, keyword)
print(f"Колонки, в яких зустрічається слово '{keyword}' : {columns_with_remote}")
print("="*50)
remote_stat = public_copy['RemoteWork'].value_counts()
print(f"Кількість респондентів на віддалений роботі (remote): {remote_stat['Remote']}")
print("\n")
#графік
plt.figure(figsize = (10, 6))
plt.pie(remote_stat,
        labels = remote_stat.index,
        autopct='%1.1f%%')
plt.title(f"Розподіл обраних форматів роботи", fontweight = 'bold', fontsize = 14)
plt.show()
```

Завдання 7. Географічний аналіз компенсації Python-розробників

```python
#берем датасет із 5 завдання з відфільтрованими Python-розробниками
print(f"Тип данних у стовпці 'ConvertedCompYearly': {python_devs_df['ConvertedCompYearly'].dtype}")
print(f"Кількість пропущених значень 'ConvertedCompYearly': {python_devs_df['ConvertedCompYearly'].isnull().sum()}")
#mean(), median() автоматично ігнорують пропущені значення, тому не видаляємо пропуски
print("="*50)
py_devs_by_country = python_devs_df.groupby('Country')['ConvertedCompYearly'].agg({'mean', 'median', 'count'}).round(2)
#видаляємо пропуски із фінальної статистики (щоб не було NaN)
py_devs_by_country_final = py_devs_by_country.dropna()
py_devs_by_country_final = py_devs_by_country_final.sort_values(by = 'median', ascending=False)
print(py_devs_by_country_final)
```

*Завдання 9. Аналіз популярності Python по віковим категоріям

```python
'''
keyword = "years"
columns_with_age = find_columns_by_value(public_copy, keyword)
print(f"Колонки, в яких зустрічається слово '{keyword}' : {columns_with_age}")
print("="*50)
'''
#вивчимо загальну статистику
age_stat = public_copy['Age'].value_counts()
#print(age_stat)
#створюємо колонку 'worked_with_python' (True/False)
public_copy['worked_with_python'] = public_copy['LanguageHaveWorkedWith'].str.contains("python", case = False, na = False)
#print(public_copy['worked_with_python'].value_counts)

py_age_stat = public_copy.groupby('Age').agg(
    all_devs = ('worked_with_python', 'count'),
    py_devs = ('worked_with_python', 'sum')
).reset_index()

py_age_stat['py_rate'] = (py_age_stat['py_devs']/py_age_stat['all_devs'])
py_age_stat['py_rate_percent'] = (py_age_stat['py_rate'] * 100).round(2)
py_age_stat.sort_values(by='py_rate_percent', ascending=False, inplace=True)
print("Аналіз популярності Python по віковим категоріям")
print("="*50)
print(py_age_stat[['Age', 'all_devs', 'py_devs', 'py_rate_percent']])
print("\n")
#візуалізуємо результати
plt.figure(figsize = (10, 6))
sns.barplot(x='Age',
            y='py_rate_percent',
            data=py_age_stat,
            color = 'purple')
plt.title("Популярність Python по віковим категоріям", fontweight = 'bold')
plt.xlabel("Вікова група", fontweight = 'bold')
plt.ylabel("Частка Python (%)", fontweight = 'bold')
plt.xticks(rotation = 45)
plt.show()
```

Висновок: чим молодша вікова група, тим вищою буде частка Python

*Завдання 10. Аналіз індустрій серед високооплачуваних віддалених працівників
